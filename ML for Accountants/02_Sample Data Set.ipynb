```python
from random import randrange
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [int(randrange(1, 501)) for i in range(1, 501)]
```

```python
y = [ (0.5 * i ** 2) + i + int(randrange(0, 50_000)) for i in x ]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
df = pd.DataFrame({
    'x': x,
    'y': y
})

df.to_csv('dataset1.csv', index=False)
```
