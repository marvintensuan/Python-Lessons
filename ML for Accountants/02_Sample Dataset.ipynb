```python
from random import randrange
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [int(randrange(1, 3_000)) for i in range(1, 3_000)]

def fn(n):
    return (25*n) + 5_000 + int(randrange(25_000))

y = list(map(fn, x))

plt.scatter(x, y, s=0.80)
plt.show()
```

```python
pd.DataFrame({
    'x': x,
    'y': y
}).to_csv('dataset1.csv', index=False)
```
