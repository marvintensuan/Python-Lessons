```python
from random import choice, randrange
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [int(randrange(1, 3_001)) for _ in range(1, 3_001)]

def fn(n):
    return int((choice(list(range(5,8)))*n) + 5_000 + randrange(3_000))

y = list(map(fn, x))

plt.scatter(x, y, s=0.80)
plt.show()
```

```python
pd.DataFrame({
    'x': x,
    'y': y
}).to_csv('dataset1.csv', index=False)
```
